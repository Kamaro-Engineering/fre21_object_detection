# Dataset visualization

Our dataset has been created by painting certain image areas with specific colors. This can be done using any standard image editing program such as GIMP. This has the advantage that all available drawing/selection tools can be used to mark areas with specific colors.

The following colors are "special" and represent a class label:
* blue (`#0000ff`): weeds
* green (`#00ff00`): maize
* red (`#ff0000`): litter

Some "tolerance" is applied. For the exact implementation, see `gazebo_screenshot_dataset.py`.

```python
dataset = GazeboScreenshotDataset(root_dir="./gazebo_data/images/")
```

```python
for sample in dataset:
    x = sample['x']
    y = sample['y']
    plt.figure()
    plt.imshow(x)
    plt.close()
    plt.figure()
    plt.imshow(y, vmin=0, vmax=3)
plt.show()
```
